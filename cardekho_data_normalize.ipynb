```python
x = 'New Car (On-Road Price) : Rs.  25.15-1.41 Crore*'
```

```python
def clean_on_road_price(x):
    amt = 0
    z = x.split(': Rs.')
    if len(z) == 1:
        if z[0].startswith('₹'):
            m = z[0].replace(',','').strip('₹ ')
            amt = float(m)
        else:
            amt = 0
    if len(z) == 2:
        z = z[1].strip()
        if "Lakh" in z:
            m = z.strip("Lakh*").strip().split("-")
            if len(m) == 2:
               avg = (float(m[0]) + float(m[1]))/2
            else:
                avg = float(m[0])
            amt = avg * pow(10,5)
        if "Crore" in z:
            m = z.strip("Crore*").strip().split("-")
            if len(m) == 2:
                if float(m[0]) > float(m[1]):
                    amt = (float(m[0])*pow(10,5) + float(m[1])*pow(10,7))/2
                else:
                    avg = (float(m[0]) + float(m[1]))/2
                    amt = avg * pow(10,7)
            else:
                amt = float(m[0]) * pow(10,7)
    return float(amt)
```
